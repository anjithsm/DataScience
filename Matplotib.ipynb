```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt
```

# **ViolinPlot**

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violin Plot of Sepal Length for Each Iris Species')
plt.show()
```

# **Swarmplot**

```python
sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title('Swarm Plot of Sepal Length for Each Iris Species')
plt.show()
```

# **Stripplot**

```python
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True)
```

# **Boxplot**

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
```

# **BarPlot**

```python
sns.barplot(x='species', y='sepal_length', data=iris, ci=None)
```

# **Countplot**

```python
sns.countplot(x='species', data=iris)
```

# **HeatMap**

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris_numeric = iris.drop(columns='species')
corr = iris_numeric.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# **Scatter**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='Set1')
plt.title('Scatter Plot of Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()
```

# **Linechart**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='species', y='sepal_length', data=iris, marker='o', hue='species')
plt.title('Line Plot of Sepal Length for Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.legend(title='Species')
plt.show()
```
